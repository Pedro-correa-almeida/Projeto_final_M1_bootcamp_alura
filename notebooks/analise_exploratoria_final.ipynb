## 2 - Leitura dos dados limpos

```python
#dados referentes às AIHs aprovadas e à população
AIH=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/AIH_aprovadas_e_populacao.csv', parse_dates=['data'],index_col=0)

#dados referentes à média de permanência dos pacientes
permanencia=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/media_permanencia.csv', parse_dates=['data'],index_col=0)

#dados referentes aos valores hospitalares gastos e valores hospitalares gastos corrigidos com o IPCA
valor=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/valor_hospitalar.csv', parse_dates=['data'],index_col=0)
```

## 3 - Funções

### 3.1 - Funções utilizadas nesse notebook

```python
@ticker.FuncFormatter
def thousand_formatter(x, pos):
    '''
    Formata os dados para cada Mil e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mil" % (x/1E3)
```

```python
@ticker.FuncFormatter
def million_formatter(x, pos):
    '''
    Formata os dados para cada Milhão e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mi" % (x/1E6)
```

### 3.2 - Funções auxiliares utilizadas em outros notebooks

## 4 - Concatenação dos dados

### 4.1 - União dos dados em apenas um DataFrame

```python
dados=pd.merge(pd.merge(AIH, permanencia, on=['UF', 'regiao', 'data']),valor, on=['UF', 'regiao', 'data'])
dados
```

### 4.2- Tratamento dos dados concatenados

```python
#criando coluna com os dados de valores corrigidos per capita 
dados['valor_corrigido_per_capita'] = dados['valor_corrigido']/ dados['pop'] 

#criando coluna com os dados de valores hospitalares per capita 
dados['valor_per_capita'] = dados['valor_hospitalar']/ dados['pop'] 
dados
```

Agora vou criar uma tabela com as informações agrupadas por região e data para comparar o comportamento ao longo do tempo dessas variáveis ao longo do tempo, por isso não utilizarei os dados de valores sem correção, pois os valores corrigidos se mostram mais reais ao longo do tempo.

```python
#agrupando os dados por região e data, somando os valores absolutos(valor gasto, valor gasto corrigido, AIHs aprovadas), 
#e fazendo a média dos estados para valores de tendência(permanência média, AIHs aprovadas por mil habitantes, valor gasto corrigido per capita, valor gasto per capita)
total_reg = dados.groupby(['regiao','data']).agg({'AIH_aprovadas':'sum', 'perm_media':'mean',
                                                 'AIH_por_mil':'mean',
                                                 'valor_corrigido':'sum', 'valor_corrigido_per_capita':'mean',}).reset_index()
total_reg
```

Aqui em vez de agrupar por data, vou agrupar os estados pelo período inteiro, e por esta razão utilizarei os valores sem correção, pois o comportamento ao longo do tempo não é um fator importante na hora de comparar esses valores absolutos.

```python
#agrupando os dados por estado , somando os valores absolutos(valor gasto, AIHs aprovadas), 
#e fazendo a média dos estados para valores de tendência(permanência média, AIHs aprovadas por mil habitantes, valor gasto per capita)
total_estado = dados.groupby(['UF', 'regiao']).agg({'AIH_aprovadas':'sum', 'perm_media':'mean',
                                        'AIH_por_mil':'mean',
                                        'valor_hospitalar':'sum','valor_per_capita':'mean'}).reset_index()
total_estado
```

## 5 - Análise Exploratória

Primeira análise que vou analisar será de gastos por estado em todo o período, nesse caso como eu não querendo ver o comportamento ao longo do tempo, usarei os dados dos valores sem correção.

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(list(reversed(sns.color_palette('Purples', 50))))

#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) 

#Criando o gráfico 
sns.barplot(x='UF', y='valor_hospitalar', data=total_estado.sort_values('valor_hospitalar',ascending=False))

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Gasto Hospitalar Total Do SUS Entre Fevereiro De 2008 E Dezembro De 2019 Por Estado', fontsize=20)
plt.xlabel('Estado', fontsize=15)
plt.ylabel('Valor Total', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=75)

#Mostrando o gráfico
plt.show()
```

Analisando os dados absolutos do período inteiro percebe se que São Paulo teve um gasto muito maior que os demais, porém São Paulo é um estado muito grande com uma população muito grande, o ideal nesse caso para descobrir os estados que mais gastam seria comparar os dados per capita. 

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette('CMRmap', 5))

#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) 

#Criando o gráfico 
sns.barplot(x='UF', y='valor_per_capita', hue='regiao', data=total_estado.sort_values(['regiao','valor_per_capita'],ascending=False),dodge=False)

#Plotando as legendas e mudando o tamanho das fontes
plt.title('Gasto Hospitalar Per Capita Do SUS Entre Fevereiro De 2008 E Dezembro De 2019 Por Estado', fontsize=20)
plt.xlabel('Estado', fontsize=15)
plt.ylabel('Valor Per Capita', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=75)
plt.legend(title='Região')

#Mostrando o gráfico
plt.show()
```

Agora com esse gráfico per capita percebemos que os estados do Sul na verdade são os estados que mais gastam per capita, eu imagino que esses estados na verdade não são tão grandes e recebem uma verba grande se comparadados com estado do Norte, e por outro lado esse gráfico mostra que na verdade a diferença per capita não parece tão grande se comparada com o valor absoluto.

Vou fazer essa mesma análise mas agora em relação às AIH(Autorização de Internação Hospitalar) aprovadas.

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(list(reversed(sns.color_palette('Purples', 50))))

#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) 

#Criando o gráfico 
sns.barplot(x='UF', y='AIH_aprovadas', data=total_estado.sort_values('AIH_aprovadas',ascending=False))

#Plotando as legendas e mudando o tamanho das fontes
ax.yaxis.set_major_formatter(million_formatter)
plt.title('Quantidade de Autorizações de Internação Hospitalar Aprovadas\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Estado', fontsize=20)
plt.xlabel('Estado', fontsize=15)
plt.ylabel('AIHs', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=75)

#Mostrando o gráfico
plt.show()
```

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette('CMRmap', 5))

#Criando figura e eixos
fig, ax = plt.subplots(figsize=(20,8)) 

#Criando o gráfico 
sns.barplot(x='UF', y='AIH_por_mil', hue='regiao', data=total_estado.sort_values(['regiao','AIH_por_mil'],ascending=False),dodge=False)

#Plotando as legendas e mudando o tamanho das fontes
plt.title('Quantidade de Autorizações de Internação Hospitalar Aprovadas\nÀ Cada Mil Habitantes Entre Fevereiro De 2008 E Dezembro De 2019 Por Estado', fontsize=20)
plt.xlabel('Estado', fontsize=15)
plt.ylabel('AIHs Por Mil Habitantes', fontsize=15)
plt.yticks(fontsize=12)
plt.xticks(fontsize=12, rotation=75)
plt.legend(title='Região')

#Mostrando o gráfico
plt.show()
```

Agora para analisar como as variáveis se relacionam vamos estudar elas ao longo do tempo e estudar se uma tem relação com a outra.

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette("Set1", 10))
    
#Criando figura e eixos
fig, ax = plt.subplots(3,1,figsize=(20,14), sharex=True) 

#Criando os gráficos 
sns.lineplot(x='data', y='AIH_aprovadas', hue='regiao', data=total_reg, ax=ax[0])
sns.lineplot(x='data', y='perm_media', hue='regiao', data=total_reg, ax=ax[1])
sns.lineplot(x='data', y='valor_corrigido', hue='regiao', data=total_reg, ax=ax[2])

#Configurando eixos, título e legendas
plt.sca(ax[0])
plt.title('Total De AIHs Aprovadas Pelo SUS\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('AIH Aprovadas', fontsize=15)
ax[0].yaxis.set_major_formatter(thousand_formatter)

plt.sca(ax[1])
plt.title('Média De Permanência No SUS\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Média De Permanência', fontsize=15)

plt.sca(ax[2])
plt.title('Total De Valor Gasto Corrigido\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Valor Corrigido', fontsize=15)
ax[2].yaxis.set_major_formatter(million_formatter)

handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[1:], labels=labels[1:], title="Região", bbox_to_anchor=(1.1, 1), fontsize=12, title_fontsize=15)
ax[1].legend('')
ax[2].legend('')

plt.sca(ax[2])
plt.xticks(pd.date_range(start=total_reg['data'].iloc[5], end =total_reg['data'].iloc[-1], freq='6M'),total_reg['data'].iloc[5::6].dt.strftime('%b/%Y'), fontsize=12, rotation=90)

#Mostrando o gráfico
plt.show()
```

Em um primeiro momento o que mais chamou minha atenção foi gráfico de média de permanência, principalmente da região Sudeste, em que tiveram meses onde a média foi fora do normal, porém se observamos nos últimos anos ela vem baixando e se aproximando das demais regiões, eu acredito que esses picos podem ter sido por algum problema na gestão dos hospitais.

Além disso, a partir da análise dos gráficos sobre o total de AIHs aprovadas, média de permanência e valor hospitalar gasto, parece que há uma relação entre essas tês variáveis, se observamos o primeiro e o último gráfico as 3 regiões que mais tiveram pessoas internadas foram as que mais gastaram. Para analisar se existe alguma relação vamos analisar a correlação dessas três variáveis com um 'heatmap'(mapa de correlação) por estado.

```python
fig, axn = plt.subplots(7, 4, sharex=False, sharey=True, figsize=(20,16))
cbar_ax = fig.add_axes([.91, .3, .03, .4])

for uf, i, ax in zip(dados['UF'].unique(), range(len(dados['UF'].unique())), axn.flat):
    corr=dados.loc[dados['UF'] == uf, ['perm_media', 'valor_corrigido', 'AIH_aprovadas']].corr()
    sns.heatmap(corr, ax=ax,
                cbar=i == 0,
                vmin=-1, vmax=1,
                cbar_ax=None if i else cbar_ax,
               cmap="RdYlBu",annot=True, fmt=".2f")
    ax.set_title(uf, fontsize=12)
    ax.set_xticklabels([],fontsize=10)
    ax.set_yticklabels(['Média De\nPermanência', 'Valor Total', 'AIHs Aprovadas'],fontsize=10, rotation=0)

axn[6][0].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'AIHs Aprovadas'],fontsize=10)  
axn[6][1].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'AIHs Aprovadas'],fontsize=10)
axn[6][2].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'AIHs Aprovadas'],fontsize=10)
axn[5][3].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'AIHs Aprovadas'],fontsize=10)
plt.delaxes(ax= axn[6][3])

fig.suptitle('Correlação Entre Total de AIHs Aprovadas, Média De Permanência\n Dos Pacientes E Total De Valor Hospitalar Gasto Pelo SUS Por Estado', fontsize=25)
plt.show()
```

A partir do grafico de correlação nota se que em cada estado existe uma relação bem diferente entre essas variáveis, porém se analisarmos bem, todos os estados apresentam uma correlação positiva entre o valor e a quantidade de AIHs aprovadas, o que faz sentido, uma vez que quanto mais verba determinado estado recebe mais internações é capaz de realizar e quanto mais internações realiza, provavelmente, maior é o gasto.

Outra informação que podemos perceber é que vários estados, como Tocantins, Paraíba, Espiríto Santo, Paraná e Distrito Federal, a média de permanência e a quantidade de AIHs aprovadas apresenta uma correlação negativa, o que eu acredito que possa influenciar nessa relação é que , provavelmente, em meses onde a média de permanência aumenta muito as vagas diminuem e, consequentemente, as internações diminuem.

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette("Set1", 10))
    
#Criando figura e eixos
fig, ax = plt.subplots(3,1,figsize=(20,14), sharex=True) 

#Criando os gráficos 
sns.lineplot(x='data', y='AIH_por_mil', hue='regiao', data=total_reg, ax=ax[0])
sns.lineplot(x='data', y='perm_media', hue='regiao', data=total_reg, ax=ax[1])
sns.lineplot(x='data', y='valor_corrigido_per_capita', hue='regiao', data=total_reg, ax=ax[2])

#Configurando título e legendas
plt.sca(ax[0])
plt.title('Média De AIHs Aprovadas À Cada Mil Habitantes Pelo SUS\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('AIHs Aprovadas Por Mil Habitantes', fontsize=15)
ax[0].yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.sca(ax[1])
plt.title('Média De Permanência No SUS\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Média De Permanência', fontsize=15)

plt.sca(ax[2])
plt.title('Média De Valor Gasto Corrigido Per Capita\nEntre Fevereiro De 2008 E Dezembro De 2019 Por Região', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Valor Corrigido Per Capita', fontsize=15)

handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[1:], labels=labels[1:], title="Região", bbox_to_anchor=(1.1, 1), fontsize=12, title_fontsize=15)
ax[1].legend('')
ax[2].legend('')

plt.sca(ax[2])
plt.xticks(pd.date_range(start=total_reg['data'].iloc[5], end =total_reg['data'].iloc[-1], freq='6M'),total_reg['data'].iloc[5::6].dt.strftime('%b/%Y'), fontsize=12, rotation=90)

#Mostrando os gráficos
plt.show()
```
