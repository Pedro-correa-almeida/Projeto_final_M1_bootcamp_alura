## 2 - Leitura dos dados limpos

```python
internacoes=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/internacoes.csv', parse_dates=['data'],index_col=0)
permanencia=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/media_permanencia.csv', parse_dates=['data'],index_col=0)
valor=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/valor_hospitalar.csv', parse_dates=['data'],index_col=0)
```

## 3 - Funções

### 3.1 - Funções utilizadas nesse notebook

```python
@ticker.FuncFormatter
def thousand_formatter(x, pos):
    '''
    Formata os dados para cada Mil e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mil" % (x/1E3)
```

```python
@ticker.FuncFormatter
def million_formatter(x, pos):
    '''
    Formata os dados para cada Milhão e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mi" % (x/1E6)
```

### 3.2 - Funções auxiliares utilizadas em outros notebooks

#### [funcoes_datasus](https://github.com/Pedro-correa-almeida/Projeto_modulo_1/blob/main/notebooks/funcoes_datasus.py)

## 4 - Concatenação dos dados

### 4.1 - União dos dados em apenas um DataFrame

```python
dados=pd.merge(pd.merge(internacoes, permanencia, on=['UF', 'regiao', 'data']),valor, on=['UF', 'regiao', 'data'])
dados
```

### 4.2- Tratamento dos dados concatenados

```python
dados['valor_corrigido_por_mil'] = dados['valor_corrigido']/ dados['pop'] * 1000
```

```python
dados
```

```python
total_reg = dados.groupby(['regiao','data']).agg({'internacoes':'sum', 'perm_media':'mean',
                                                 'valor_hospitalar':'sum', 'internacao_por_mil':'mean',
                                                 'valor_corrigido':'sum', 'valor_corrigido_por_mil':'mean'}).reset_index()
total_reg
```

## 5 - Análise Exploratória

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette("Set1", 10))
    
#Criando figura e eixos
fig, ax = plt.subplots(3,1,figsize=(20,14), sharex=True) 

#Criando o gráfico 
sns.lineplot(x='data', y='internacoes', hue='regiao', data=total_reg, ax=ax[0])
sns.lineplot(x='data', y='perm_media', hue='regiao', data=total_reg, ax=ax[1])
sns.lineplot(x='data', y='valor_corrigido', hue='regiao', data=total_reg, ax=ax[2])

plt.sca(ax[0])
plt.title('Total De Internações Pelo SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Internações', fontsize=15)
ax[0].yaxis.set_major_formatter(thousand_formatter)

plt.sca(ax[1])
plt.title('Média De Permanência No SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Média De Permanência', fontsize=15)

plt.sca(ax[2])
plt.title('Total De Valor Gasto Corrigido Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Valor Corrigido', fontsize=15)
ax[2].yaxis.set_major_formatter(million_formatter)

handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[1:], labels=labels[1:], title="Região", bbox_to_anchor=(1.1, 1), fontsize=12, title_fontsize=15)
ax[1].legend('')
ax[2].legend('')

plt.sca(ax[2])
plt.xticks(pd.date_range(start=total_reg['data'].iloc[5], end =total_reg['data'].iloc[-1], freq='6M'),total_reg['data'].iloc[5::6].dt.strftime('%b/%Y'), fontsize=12, rotation=90)
#Mostrando o gráfico
plt.show()
```

A partir da análise dos gráficos sobre o total de internações, média de permanência e valor hospitalar gasto, parece que há uma relação entre essas tês variáveis, se observamos o primeiro e o último gráfico as 3 regiões que mais tiveram pessoas internadas foram as que mais gastaram. Para analisar se existe alguma relação vamos analisar essa correlação com um 'heatmap'(mapa de correlação) pela região.

```python
fig, axn = plt.subplots(7, 4, sharex=False, sharey=True, figsize=(20,16))
cbar_ax = fig.add_axes([.91, .3, .03, .4])

for uf, i, ax in zip(dados['UF'].unique(), range(len(dados['UF'].unique())), axn.flat):
    corr=dados.loc[dados['UF'] == uf, ['perm_media', 'valor_corrigido', 'internacoes']].corr()
    sns.heatmap(corr, ax=ax,
                cbar=i == 0,
                vmin=-1, vmax=1,
                cbar_ax=None if i else cbar_ax,
               cmap="RdYlBu",annot=True, fmt=".2f")
    ax.set_title(uf, fontsize=12)
    ax.set_xticklabels([],fontsize=10)
    ax.set_yticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=10, rotation=0)

axn[6][0].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=10)  
axn[6][1].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=10)
axn[6][2].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=10)
axn[5][3].set_xticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=10)
plt.delaxes(ax= axn[6][3])

fig.suptitle('Correlação Entre Total de Internações, Média De Permanência\n Dos Pacientes E Total De Valor Hospitalar Gasto Pelo SUS Por Estado', fontsize=25)
plt.show()
```

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette("Set1", 10))
    
#Criando figura e eixos
fig, ax = plt.subplots(3,1,figsize=(20,14), sharex=True) 

#Criando o gráfico 
sns.lineplot(x='data', y='internacao_por_mil', hue='regiao', data=total_reg, ax=ax[0])
sns.lineplot(x='data', y='perm_media', hue='regiao', data=total_reg, ax=ax[1])
sns.lineplot(x='data', y='valor_corrigido_por_mil', hue='regiao', data=total_reg, ax=ax[2])

plt.sca(ax[0])
plt.title('Média De Internações Por Mil Habitantes Pelo SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Internações', fontsize=15)
ax[0].yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.sca(ax[1])
plt.title('Média De Permanência No SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Média De Permanência', fontsize=15)

plt.sca(ax[2])
plt.title('Média De Valor Gasto Corrigido Por Mil Habitantes Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Valor Corrigido', fontsize=15)
ax[2].yaxis.set_major_formatter(thousand_formatter)

handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[1:], labels=labels[1:], title="Região", bbox_to_anchor=(1.1, 1), fontsize=12, title_fontsize=15)
ax[1].legend('')
ax[2].legend('')

plt.sca(ax[2])
plt.xticks(pd.date_range(start=total_reg['data'].iloc[5], end =total_reg['data'].iloc[-1], freq='6M'),total_reg['data'].iloc[5::6].dt.strftime('%b/%Y'), fontsize=12, rotation=90)
#Mostrando o gráfico
plt.show()
```

Dessa vez observando os valores em relação à mil habitantes dos estados, percebemos que na quantidade de internações não varia tanto nas regiões, o que faz sentido, já que a população não interfere nessa análise não deve haver tanta diferença entre as regiões, porém dessa vez analisando os valores gastos percebe que a média de gastos dos estados do sul em cada mil habitantes 'bem superior que as outras, e por outro lado a região norte está bem abaixo, eu imagino que isso pode ser pelo fato dos estados da região sul, mesmo não tendo uma população tão grande, recebem altas verbas para saúde.
