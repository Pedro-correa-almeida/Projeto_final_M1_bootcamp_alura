## Limpeza dos dados referentes ao ipca mensal

```python
ipca.head(5)
```

Primeira coisa a se fazer vai ser renomear as colunas para entender esses dados, olhando o DataFrame percebe se que a primeira indica o ano, a segunda o mês e a terceira o ipca acumulativo, que de acordo com a fonte de dados está tendo como referência dezembro de 1993.

```python
ipca.columns=['Ano', 'Mês', 'ipca_93']
ipca.head(25)
```

Exibindo as 25 primeiras linhas do DataFrame percebe se que a coluna referente ao ano apresentam muitos dados nulos, eu acredito que o ano correspondente é a linha mais próxima que não tem um valor nulo, ou seja, todos os dados entre 1994 e 1995 são do ano 1994.

Para preencher os dados nulos dessa coluna ano vou usar o método de preencher com os valores nulos com o valor anterior mais próximo diferente de nulo.

```python
ipca['Ano'] = ipca['Ano'].fillna(method='ffill')
ipca.head(25)
```

```python
ipca.sample(25, random_state=422)
```

Se pegarmos linhas aleatórias do DataFrame percebe se que algumas linhas na coluna 'Ano' e da coluna 'ipca_93' na verdade estão com texto ou com valores nulos, mas são informações que não serão utilizadas na análise e por isso serão deletadas, para isso vou criar um for para tentar converter o valor do 'ano' e do 'ipca_93' para um inteiro, e se não conseguir essa linha será eliminada.

```python
for i in range(len(ipca['Ano'])):
    try:
        int(ipca['Ano'][i])
        int(ipca['ipca_93'][i])
    except:
        ipca = ipca.drop(i)
```

Para verificar se deu certo vou tentar converter a coluna 'Ano' para inteiro e a coluna 'ipca_93' para float

```python
ipca['Ano'] = ipca['Ano'].astype('int64')
ipca['ipca_93'] = ipca['ipca_93'].astype('float64')
ipca
```

```python
ipca.info()
```

Agora as colunas estão com todas com tipo correto e não apresentam valor nulo.

Após isso vou limitar os dados para o padrão estabelecido, entre fevereiro de 2008 e dezembro 2019 trasnformando em tipo data, e calcular o ipca em relação ao último mês(dezembro de 2019)

Para transformar as colunas 'Ano' e 'Mês' em uma só do tipo data utilizarei a função tranforma_date do arquivo .py(funcoes_datasus)

```python
help(tranforma_date)
```

```python
#limitando  os dados entre 2008 e 2019
ipca = ipca[(ipca['Ano'] >= 2008) & (ipca['Ano'] <= 2019)]

#Calculando o ipca em relação ao último mês
ipca['ipca_93'] = ipca['ipca_93']/ipca['ipca_93'].iloc[-1] 

#Juntando a coluna 'Ano' e 'Mês' e transformando no tipo data
ipca['data'] = tranforma_date(ipca['Ano'].astype(str) +'/'+ ipca['Mês'].str.capitalize())
ipca = ipca.drop(columns=['Ano', 'Mês'])
ipca
```

Agora que temos o DataFrame do ipca no mesmo padrão dos dados de valores podemos juntar os dois.

```python
#Juntando os dois DataFrames
dados = dados.merge(ipca, on='data')

#Caluclando o valor pela inflação em relação ao último mês
dados['valor_corrigido'] =dados['valor_hospitalar'] / dados['ipca_93']
dados['valor_corrigido'] = dados['valor_corrigido'].astype('float64')

#Eliminando a coluna do ipca
dados=dados.drop(columns='ipca_93')

dados
```

Agora vamos visualizar como ficou esse valor corrigido.

```python
plot_grafico_linha(dados, y='valor_corrigido', title='Valor Hospitalar Corrigido Entre 2008 E 2020 Por Estado', legend_title='Estado', ylabel='Valor')
```

Olhando por estado nota se uma diferença muito grande com o gráfico anterior, antes parecia que os estados estavam em uma constante cresecente durante os anos, mas se corrigirmos esse valor pelo ipca de cada mês, percebe se que na verdade a muitos estados estão diminuindo seus gastos.

```python
#agrupando por região e data
dados_reg = dados.groupby(['regiao', 'data']).sum().reset_index()
dados_reg                       
```

```python
plot_grafico_linha(dados_reg, y='valor_corrigido', hue='regiao', 
                   title='Valor Hospitalar Corrigido Entre 2008 E 2020 Por Região', legend_title='Região', ylabel='Valor')
```

```python
plot_grafico_linha(dados_reg, y='valor_hospitalar', hue='regiao', 
                   title='Valor Hospitalar Entre Fevereiro De 2008 E Dezembro De 2019 Por Região',legend_title='Região', ylabel='Valor')

plot_grafico_linha(dados_reg, y='valor_corrigido', hue='regiao', 
                    title='Valor Hospitalar Corrigido Entre 2008 E 2020 Por Região', legend_title='Região', ylabel='Valor')
```

Observando agora os valores totais por região, parece que a maioria das regiões não apresentou grande mudança nos gastos de valores corrigidos, apenas o Sudeste que nesse intervalo de tempo sempre foi o que mais gastou e sofreu uma redução nos gastos nos últimos anos, e se compararmos com os dados antes de serem corrigidos parecem fazer mais sentido na linha do tempo que não é tão grande para se ter uma aumento desses como é visto no primeiro gráfico.

Por estas razões, eu imagino que na hora de comparar o comportamento dos valores ao longo do tempo seja melhor usar esse valor corrigido, porém para saber dos valores absolutos sem levar em conta o tempo o valor real talvez seja uma melhor escolha, por isso vou deixar as duas colunas.
