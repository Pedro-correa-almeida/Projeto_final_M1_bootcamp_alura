# Análise e Limpeza dos dados relacionados a média de permanência dos pacientes internados pelo Sus 

## Importações dos pacotes

```python
import pandas as pd
from funcoes_datasus import limpa_dados, plot_grafico_linha 
```

## Dados

```python
dados = pd.read_csv("C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_brutos/A111755189_28_143_208_media_perm.csv", encoding="ISO-8859-1", 
                    skiprows = 3, sep=";", skipfooter=12,
                    na_values='-',
                    thousands=".", decimal=",", engine="python")
```

Os dados foram coletados do datasus e dizem respeito a média de permanência dos pacientes internados pelo SUS processadas a partir de Janeiro de 2008 por mês e por Unidade Federativa, a fim de causar falsas impressões e ter certeza que os dados são reais e estão atualizados, os dados antes de janeiro de 2008(quando foram processados) serão removidos,o mês de janeiro de 2008 será removido também pois pode haver algum número equivocado.  

Além disso, segundo a própia fonte de dados:'Dados referentes aos últimos seis meses, sujeitos a atualização.', então esses dados desatualizados serão removidos também, e por último a coluna total será removida a fim de melhorar a visualização.

```python
dados.mean()
```

Podemos perceber que realmente os dados antes de fevereiro de 2008 estão fora do padrão, com uma variação muito grande, sendo uma das justificativas para o padrão de limpeza dos dados estabelecido no escopo.

Para deixar os dados no padrão utilizarei a função limpa_dados do arquivo funcoes_datasus.py:

https://github.com/Pedro-correa-almeida/Projeto_modulo_1/blob/main/notebook/funcoes_datasus.py

```python
help(limpa_dados)
```

```python
dados = limpa_dados(dados, 'perm_media')
```

```python
dados
```

```python
dados.isna().sum()
```

```python
dados.dtypes
```

Agora percebe se que os dados estão em um um formato mais ideal para concatenar com outras bases de dados, além disso os dados estão limitados entre fevereiro de 2008 e dezembro de 2019, e a coluna 'data' é do tipo datetime64[ns] representando data.

Para visualizar esses dados ao longo do tempo por região utilizarei a função plot_grafico_linha do arquivo funcoes_datasus.py

```python
help(plot_grafico_linha)
```

```python
#Plotando o gráfico
plot_grafico_linha(dados, 'perm_media', hue='UF',title='Média De Permanência Dos Pacientes Internados Pelo SUS \nEntre Fevereiro De 2008 E Agosto De 2020 Por Estado',
                   legend_title='Estado', ylabel='Média De Permanência')
```

Nesse gráfico não fica clara essa média de permanência dos pacientes por estados, pois são muitos estados e algum deles parecem ter uma média muito maior em determinados meses.

```python
dados_reg = dados.groupby(['regiao', 'data']).mean().reset_index()
dados_reg
```

```python
plot_grafico_linha(dados_reg, 'perm_media', hue='regiao', title='Média De Permanência Dos Pacientes Internados Pelo SUS \nEntre Fevereiro De 2008 E Agosto De 2020 Por Região',
                   ylabel='Média De Permanência', legend_title='Região')
```

Podemos perceber qua a maior parte dos estados tem uma média de permanência bem próxima, mas alguns tem uma variação muito alta em determinadas épocas, esses dados serão mais explorados no notebook com todas as bases de dados juntas, além disso percebe se que em anos anteriores tinha se uma média muito maior, principalmente no Sudeste, o que pode indicar uma melhora na gestão, pois em alguns meses a média de permanência foi muito acima do normal.
