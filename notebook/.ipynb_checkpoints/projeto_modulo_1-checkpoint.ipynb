## Importações dos pacotes

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker
import seaborn as sns
import numpy as np
```

```python
internacoes=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/internacoes.csv', parse_dates=['data'],index_col=0)
permanencia=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/media_permanencia.csv', parse_dates=['data'],index_col=0)
valor=pd.read_csv('C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/valor_hospitalar.csv', parse_dates=['data'],index_col=0)
```

```python
@ticker.FuncFormatter
def thousand_formatter(x, pos):
    '''
    Formata os dados para cada Mil e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mil" % (x/1E3)
```

```python
@ticker.FuncFormatter
def million_formatter(x, pos):
    '''
    Formata os dados para cada Milhão e mostra duas casas decimais depois da vírgula
    '''
    return "%.2f Mi" % (x/1E6)
```

```python
dados=pd.merge(pd.merge(internacoes, permanencia, on=['UF', 'regiao', 'data']),valor, on=['UF', 'regiao', 'data'])
dados
```

```python
dados['valor_corrigido_por_mil'] = dados['valor_corrigido']/ dados['pop'] * 1000
```

```python
dados
```

```python
total_reg = dados.groupby(['regiao','data']).agg({'internacoes':'sum', 'perm_media':'mean',
                                                 'valor_hospitalar':'sum', 'internacao_por_mil':'mean',
                                                 'valor_corrigido':'sum', 'valor_corrigido_por_mil':'mean'}).reset_index()
total_reg
```

```python
#Mudando o estilo de fundo
sns.set_style('darkgrid')
    
#Criando uma palheta de cores
sns.set_palette(sns.color_palette("Set1", 10))
    
#Criando figura e eixos
fig, ax = plt.subplots(3,1,figsize=(20,14), sharex=True) 

#Criando o gráfico 
sns.lineplot(x='data', y='internacoes', hue='regiao', data=total_reg, ax=ax[0])
sns.lineplot(x='data', y='perm_media', hue='regiao', data=total_reg, ax=ax[1])
sns.lineplot(x='data', y='valor_corrigido', hue='regiao', data=total_reg, ax=ax[2])

plt.sca(ax[0])
plt.title('Total De Internações Pelo SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Internações', fontsize=15)
ax[0].yaxis.set_major_formatter(thousand_formatter)

plt.sca(ax[1])
plt.title('Média De Permanência No SUS Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Média De Permanência', fontsize=15)

plt.sca(ax[2])
plt.title('Total De Valor Gasto Corrigido Por Região Entre 2008 E 2020', fontsize=20)
plt.yticks(fontsize=12)
plt.ylabel('Valor Corrigido', fontsize=15)
ax[2].yaxis.set_major_formatter(million_formatter)

handles, labels = ax[0].get_legend_handles_labels()
ax[0].legend(handles=handles[1:], labels=labels[1:], title="Região", bbox_to_anchor=(1.1, 1), fontsize=12, title_fontsize=15)
ax[1].legend('')
ax[2].legend('')

plt.sca(ax[2])
plt.xticks(pd.date_range(start=total_reg['data'].iloc[5], end =total_reg['data'].iloc[-1], freq='6M'),total_reg['data'].iloc[5::6].dt.strftime('%b/%Y'), fontsize=12, rotation=90)
#Mostrando o gráfico
plt.show()
```

```python
fig, axn = plt.subplots(2, 3, sharex=False, sharey=True, figsize=(20,12))
cbar_ax = fig.add_axes([.91, .3, .03, .4])

for reg, i, ax in zip(total_reg['regiao'].unique(), range(6), axn.flat):
    corr=total_reg.loc[total_reg['regiao'] == reg, ['perm_media', 'valor_corrigido', 'internacoes']].corr()
    sns.heatmap(corr, ax=ax,
                cbar=i == 0,
                vmin=-1, vmax=1,
                cbar_ax=None if i else cbar_ax,
               cmap="RdYlBu",annot=True, fmt=".2f")
    ax.set_title(reg, fontsize=15)
    ax.set_xticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=12)
    ax.set_yticklabels(['Média De\nPermanência', 'Valor Total', 'Internações'],fontsize=12)

plt.delaxes(ax= axn[1][2])
fig.suptitle('Correlação Entre Total de Internações, Média De Permanência\n Dos Pacientes E Total De Valor Hospitalar Gasto Pelo SUS Por Região', fontsize=25)
plt.show()
```
