# Limpeza Dados população Dos Estados

```python
import pandas as pd
import re
```

```python
dados_pop = pd.read_excel('C:/Users/pedro/Documents/Bootcamp_alura/Modulo_1/dados/serie_2001_2020_TCU.xls', skiprows=4, skipfooter=10)
```

```python
dados_pop
```

```python
dados_pop = dados_pop.rename(columns={'Unidades da Federação':'UF'})
index_drop = [i for i in range(2,len(dados_pop)) if re.match(r'Região.*',dados_pop.loc[i,'UF'])]
dados_pop = dados_pop.drop(index=index_drop)
dados_pop = dados_pop.drop([0,1])
```

```python
dados_pop = pd.melt(dados_pop, id_vars='UF', var_name='ano', value_name='pop')
dados_pop = dados_pop[(dados_pop['ano'] >=2008) & (dados_pop['ano']<=2019) ]
dados_pop
```

```python
dados_pop.to_csv("C:/Users/pedro/Documents/projeto_modulo_1_bootcamp2_alura/dados_limpos/populacao_ano.csv")
```
